```python
import pandas as pd
```

```python
df_population_raw = pd.read_csv('population_total.csv')
```

```python
df_population_raw
```

# Making a Pivot Table
.pivot(): Returns reshaped DataFrame organized by given index / column values ("pivot without aggregation")

```python
# dropping null(na) values
df_population_raw= df_population_raw.dropna()
df_population_raw
```

```python
#making a pivot table
df_pivot = df_population_raw.pivot(index ="year", columns = "country", values = "population")
df_pivot
```

```python
#selecting some countries
df_pivot = df_pivot[['United States', 'India', 'China', 'Indonesia', 'Brazil']]
```

```python
#showing pivot table
df_pivot
```

# Data Visulization

## Lineplot

```python
df_pivot.plot(kind= "line", xlabel = "Years", ylabel= "Population", title = "Population (1995-2020)", figsize= (8,4))

import matplotlib.pyplot as mt   #by usimg this we can save graph/figure into other file
mt.savefig('test.png')
```

```python
df_pivot.to_excel('pivot_table.xlsx')    #this convert  the csv file itno  excel file
```

## Barplot

```python
#selecting one year only (2020)
df_pivot_2020 = df_pivot[df_pivot.index.isin([2020])]
df_pivot_2020
```

```python
#transpose dataframe from row to column  (needed for barplot)
df_pivot_2020=df_pivot_2020.T
df_pivot_2020
```

```python
df_pivot_2020.plot(kind = "bar", xlabel = "country", ylabel= "Population", title = "Population graph(2020)", figsize= (8,4), color = "aqua")
```

## Barplot with n variable

```python
df_pivot_sample = df_pivot[df_pivot.index.isin([1980, 1990, 2020, 2010, 2020])]
df_pivot_sample
```

```python
df_pivot_sample.plot(kind = "bar", xlabel = "years", ylabel = "population", title = "Population graph", figsize= (8,4))
```
